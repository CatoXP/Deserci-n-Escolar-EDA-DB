```python
import pandas as pd
```

```python
df = pd.read_csv("/content/respuestas_desercion.csv", encoding='latin1')
```

```python
df.head(20)
```

```python
def calcular_riesgo(row):
    puntaje = 100

    # Económico (25 pts)
    if '$3,000' in row.iloc[0]: puntaje -= 10
    elif '$5,000' in row.iloc[0]: puntaje -= 6
    if 'tiempo completo' in row.iloc[1]: puntaje -= 8
    if 'No tengo' in row.iloc[2]: puntaje -= 5

    # Académico (30 pts)
    if 'Menos de 8' in row.iloc[5]: puntaje -= 8
    if 'Muy bajo' in row.iloc[6] or 'Bajo' in row.iloc[6]: puntaje -= 7
    if 'Reprobé 3' in row.iloc[9] or 'Reprobé 4' in row.iloc[9]: puntaje -= 4

    # Psicosocial (25 pts)
    if 'Mal' in row.iloc[10]: puntaje -= 8
    if 'No tengo' in row.iloc[11] or 'Casi no' in row.iloc[11]: puntaje -= 6
    if 'Bajo' in row.iloc[14]: puntaje -= 3

    # Vocacional (12 pts)
    if 'influencia' in row.iloc[15] or 'moda' in row.iloc[15]: puntaje -= 4
    if 'Nada' in row.iloc[16] or 'Poco' in row.iloc[16]: puntaje -= 3
    if 'Inseguro' in row.iloc[17]: puntaje -= 3

    # Institucional (8 pts)
    if 'oponen' in row.iloc[4]: puntaje -= 3
    if 'No conozco' in row.iloc[22]: puntaje -= 2
    if 'Poco probable' in row.iloc[24] or 'Muy poco' in row.iloc[24]: puntaje -= 3

    # Categoría
    if puntaje >= 75: cat = 'BAJO'
    elif puntaje >= 50: cat = 'MEDIO'
    elif puntaje >= 25: cat = 'ALTO'
    else: cat = 'MUY ALTO'

    return puntaje, cat
```

```python
resultados = []
for i, row in df.iterrows():
    pts, cat = calcular_riesgo(row)
    resultados.append({'Estudiante': i+1, 'Puntaje': pts, 'Riesgo': cat})

res_df = pd.DataFrame(resultados)
```

```python
print("\n=== CLASIFICACIÓN DE RIESGO ===\n")
print(res_df['Riesgo'].value_counts().sort_index())
print(f"\nPuntaje promedio: {res_df['Puntaje'].mean():.1f}")
print("\nPrimeros 10 estudiantes:")
print(res_df.head(43))
```

```python
res_df.to_csv('riesgo_desercion.csv', index=False)
print("\n✓ Guardado en: riesgo_desercion.csv")
```
